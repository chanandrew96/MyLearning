```python
import time
import keras_cv
from tensorflow import keras
import matplotlib.pyplot as plt
```

```python
# Construct model
model = keras_cv.models.StableDiffusion(img_width=512, img_height=512)
```

```python
def plot_images(images):
  plt.figure(figsize=(20, 20))
  for i in range(len(images)):
    ax =plt.subplot(1, len(images), i + 1)
    plt.imshow(images[i])
    plt.axis("off")
```

```python
images = model.text_to_image("photograph of a horse running to moon", batch_size=3)
plot_images(images)
```

```python
images = model.text_to_image("Cute bear asking for hug", batch_size=3)
plot_images(images)
```
